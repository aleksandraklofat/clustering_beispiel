# Clustering Example

```python
# from sklearn.datasets import make_classification
# generates random classification

from sklearn.datasets import make_classification
from matplotlib import pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

```python
plt.title("Two informative features, one cluster per class", fontsize="large")

X, Y = make_classification(n_features=2, n_redundant=0)
plt.scatter(X[:, 0], X[:, 1], marker="o", c=Y, s=25, edgecolor="k");
```

```python
X = StandardScaler().fit_transform(X)
```

## K-Means

```python
Y_pred_kmeans = KMeans(n_clusters=2).fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y_pred_kmeans);
```

## DBSCAN

```python
Y_pred_dbscan = DBSCAN(min_samples=3).fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1],c=Y_pred_dbscan);
```
