```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

```python
X, Y = make_circles(n_samples=800, factor=0.3, noise=0.1)
```

```python
X = StandardScaler().fit_transform(X)
```

## K-Means

```python
Y_pred_kmeans = KMeans(n_clusters=2).fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1],s=8,lw=1, c=Y_pred_kmeans, cmap='magma');
```

## DBSCAN

```python
Y_pred_dbscan = DBSCAN(eps=0.3, min_samples=2).fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], s=8,lw=1,c=Y_pred_dbscan, cmap='magma');
```
